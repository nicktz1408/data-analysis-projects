### Linear Kernel

### RBF kernel
Ακολουθούμε παρόμοια στρατιγική fine tuning με τα polynomial kernel μοντέλα, μόνο που εδώ θα πρέπει να έχουμε 3D grid search προσθέτοντας το gamma.

```python
for gamma in [100, 10, 1, 0.1, 0.01, 0.001]:
    for C in [0.001, 0.01, 0.1, 1, 10, 100, 1000]:
        for epsilon in [0.001, 0.01, 0.1, 1, 10, 100, 1000]:
            model = SVR(kernel='rbf', C=C, epsilon=epsilon, gamma=gamma)
            model.fit(X_train, y_train)

            y_pred = model.predict(X_val)

            print(f'MAE γ={gamma}, C={C}, ε={epsilon}: {mean_absolute_error(y_val, y_pred)}')
```

```python
# MAE γ=0.1, C=1000, ε=1: 3.219836232402966
for gamma in [.08, .09, 1, 1.1, 1.2]:
    for C in [800, 100, 1200, 1400]:
        for epsilon in [.8, .9, 1, 1.1, 1.2]:
            model = SVR(kernel='rbf', C=C, epsilon=epsilon, gamma=gamma)
            model.fit(X_train, y_train)

            y_pred = model.predict(X_val)

            print(f'MAE γ={gamma}, C={C}, ε={epsilon}: {mean_absolute_error(y_val, y_pred)}')
```

```python
# MAE γ=0.09, C=1400, ε=1.1: 3.1967806334584967
for gamma in [.09]:
    for C in [1400, 1500, 1600, 170]:
        for epsilon in [1.1, 1.2]:
            model = SVR(kernel='rbf', C=C, epsilon=epsilon, gamma=gamma)
            model.fit(X_train, y_train)

            y_pred = model.predict(X_val)

            print(f'MAE γ={gamma}, C={C}, ε={epsilon}: {mean_absolute_error(y_val, y_pred)}')
```

Μετά από έλεγχο, παίρνω καλά αποτελέσματα για `γ=0.09, C=1600, ε=1.1` που αποδίδουν ΜΑΕ 3.1814. Δοκιμάζω το παραπάνω μοντέλο στα test data για να το συγκρίνω με τα υπόλοιπα:
